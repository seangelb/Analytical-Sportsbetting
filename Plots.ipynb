```python
import data_clean
import algorithms
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = data_clean.do(data_clean)
df = algorithms.ml_probabilities(algorithms,df)
```

```python
params = {"ytick.color" : "w",
          "xtick.color" : "w",
          "axes.labelcolor" : "w",
          "axes.edgecolor" : "w",
          "axes.edgecolor" : "w"}
plt.rcParams.update(params)

margin = df['total_implied_prob'] - 100
total = df['total_implied_prob']
edge = (margin / total) * 100
sns.kdeplot(data=edge, color='r', shade=True)
plt.title("Density of Sportsbooks' EV in NFL Moneyline", color="w")
plt.xlabel("EV % ")
#https://www.investopedia.com/articles/dictionary/042215/understand-math-behind-betting-odds-gambling.asp
```

```python
import pandas as pd
import matplotlib.pyplot as plt

params = {"ytick.color" : "w",
          "xtick.color" : "w",
          "axes.labelcolor" : "w",
          "axes.edgecolor" : "w",
          "axes.edgecolor" : "w"}
plt.rcParams.update(params)

chart = pd.read_csv('test_v15.csv')
chart['Year'] = chart['Year'].apply(lambda x: int(x))

ax = chart.plot.bar(x='Year', y=['Amount', 'Amount_EV'])
ax.set_xlabel("Year", fontsize=12)
ax.set_ylabel("Amount", fontsize=12)
plt.title("NFL ML EV > 0.07, Max Spread: -15, Starting Amt $1000", color="w")
plt.show()

```
